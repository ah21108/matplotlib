# Matplotlib Line Graph

## Available Defined Styles
Running the command "plt.style.available" will bring up list of predefined plot styles.  

To use, set plt.style.use(). If you set any subsequent attributes in the plt.plot box it should override the formatted style. Defined styles here: <html>https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html</html>
            
Popular ones are fivethirtyeight, ggplot, and seaborn(series)          

## Labels and Legends
See example below.  Including label= will automatically trigger legend graphic, rather than having to set/order legend separately.  

Include label in plt.plot() command; separate plt.legend() command will add the legend to graphic.


## Colors
### Base Colors
Blue: b; Green: g; Red: r; Cyan: c; Magenta: m; Yellow: y; Black: k; White: w
 
### Hex Colors
Can reference color by pound-sign and the hexcode

    

## Line Styles and Marks 
Named Linestyles are solid, dotted, dashed, dashdot.  Include linestyle in plot command
Others can be defined/parameterized, found here: <html>https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html</html>

<body>Common markers:<br>

4: tri_right<br>

+: plus<br>
x: x<br>

D: diamond<br>

More found here: <html>https://matplotlib.org/stable/api/markers_api.html</html>
